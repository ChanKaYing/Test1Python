```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'coffee.csv'  # Update this with the correct path to the coffee.csv file
df = pd.read_csv(file_path)

# Group by 'product_category' and 'product_type', and calculate the total orders
grouped_df = df.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()

# Rename the columns to match the requested output format
grouped_df.columns = ['category', 'product', 'total_order']

# Display the result
print(grouped_df)

# Plot the data
plt.figure(figsize=(12, 8))
for category in grouped_df['category'].unique():
    subset = grouped_df[grouped_df['category'] == category]
    plt.bar(subset['product'], subset['total_order'], label=category)

plt.xlabel('Product')
plt.ylabel('Total Orders')
plt.title('Total Orders by Product and Category')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```
